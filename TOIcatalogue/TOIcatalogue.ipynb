```python
import pandas as pd
```

```python
data = pd.read_csv('csv-file-toi-catalog.csv')
print(data.shape)
data.head()
```

```python
data1 = data[data['Orbital Period Value'] > 500]
print(data1.shape)
data1.head()
```

```python
#maybe redundant?
# df = data1[data1['Public Comment'].str.contains('period')]
# print(df.shape)
# df.head()
```

```python
#Zenith dec 52.3Â° down to +10Â°
df = data1[(data1['TIC Declination'] > 10) & (data1['TIC Declination'] < 90)]
print(df.shape)
df.head()
```

```python
#16hrs to 23.25hrs - 240.00 - 360.00
df = df[~((df['TIC Right Ascension'] > 35) & (df['TIC Right Ascension'] < 250))]
df = df.reset_index(drop=True)
print(df.shape)
df.head()
```

```python
df.to_csv('duotargets.csv')
```
