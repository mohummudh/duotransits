```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.table import Table
import lightkurve as lk
import pandas as pd
%matplotlib inline
```

```python
#accessing the lightcurves.
sec15 = 'hlsp_tess-spoc_tess_phot_0000000276376289-s0015_tess_v1_lc.fits'
sec55 = 'hlsp_tess-spoc_tess_phot_0000000276376289-s0055_tess_v1_lc.fits'
sec56 = 'hlsp_tess-spoc_tess_phot_0000000276376289-s0056_tess_v1_lc.fits'
```

```python
#file information
fits.info(sec15)
```

```python
#checking the data columns
fits.getdata(sec15, ext = 1).columns
```

```python
%matplotlib inline
#extracting the data for this TESS sector: time, flux, flux error.
with fits.open(sec15, mode='readonly', byteorder='big') as stats:
    time0 = stats[1].data['TIME']
    pdsap0 = stats[1].data['PDCSAP_FLUX']
    pdsap_err0 = stats[1].data['PDCSAP_FLUX_ERR']

#removing NaN time values    
remove = np.where(np.isnan(time0))[0]
mask = np.ones(time0.shape, dtype=bool)
mask[remove] = False

time0 = time0[mask]
pdsap0 = pdsap0[mask]
pdsap_err0 = pdsap_err0[mask]

#creating a LightKurve object and plotting the data from this TESS sector.
lc0 = lk.LightCurve(time=time0, flux=pdsap0, flux_err=pdsap_err0)

#normalising and flattening the lightcurve
lc0 = lc0.normalize().flatten()
lc0.scatter()
```

```python
#extracting the data for this TESS sector: time, flux, flux error.
with fits.open(sec55, mode='readonly') as stats:
    time1 = (stats[1].data['TIME'])
    pdsap1 = (stats[1].data['PDCSAP_FLUX'])
    pdsap_err1 = (stats[1].data['PDCSAP_FLUX_ERR'])

#removing NaN time values  
remove = np.where(np.isnan(time1))[0]
mask = np.ones(time1.shape, dtype=bool)
mask[remove] = False

time1 = time1[mask]
pdsap1 = pdsap1[mask]
pdsap_err1 = pdsap_err1[mask]

#creating a LightKurve object and plotting the data from this TESS sector.
lc1 = lk.LightCurve(time=time1, flux=pdsap1, flux_err=pdsap_err1)

#normalising and flattening the lightcurve
lc1 = lc1.normalize().flatten()
lc1.scatter()
```

```python
#extracting the data for this TESS sector: time, flux, flux error.
with fits.open(sec56, mode='readonly') as stats:
    time2 = (stats[1].data['TIME'])
    pdsap2 = (stats[1].data['PDCSAP_FLUX'])
    pdsap_err2 = (stats[1].data['PDCSAP_FLUX_ERR'])

#removing NaN time values  
remove = np.where(np.isnan(time2))[0]
mask = np.ones(time2.shape, dtype=bool)
mask[remove] = False

time2 = time2[mask]
pdsap2 = pdsap2[mask]
pdsap_err2 = pdsap_err2[mask]

#creating a LightKurve object and plotting the data from this TESS sector.
lc2 = lk.LightCurve(time=time2, flux=pdsap2, flux_err=pdsap_err2)

#normalising and flattening the lightcurve
lc2 = lc2.normalize().flatten()
lc2.scatter()
```

```python
#combing the three TESS sectors into one lightcurve

%matplotlib inline
lc = lc0.append([lc1,lc2])
lc.scatter()
```

```python
#getting the data as individual arrays

time = lc.time.jd
flux = lc.flux.value
flux_err = lc.flux_err.value
```

```python
#information about the transits in the data

epoch = 1716.9264333 #input epoch from the TOI Catalogue
period = 1084.2279131 #input period from the TOI Catalogue

t1 = epoch #transit 1
t2 = epoch + period #transit 2

#PLOTTING TRANSIT POSITIONS ON NORMALISED + FALTTENED DATA

fig, ax = plt.subplots()

ax.plot(time, flux, 'ko')
#ax.set_xlim(t2 - 0.5, t2 + 0.5)
ax.axvline(x=t1, color="red")
ax.axvline(x=t2, color="blue")
```

```python
#SETTING TRANSIT PARAMETERS + REPLOTTING

# Parameters that you need to set

t_zero_1 = t1
t_zero_2 = t2

#maximum orbital period allowed for the planet 
Pmax = np.abs(t2 - t1)

width = (3.518/24) #width from TOI catalogue (hours -> days)
depth =  12356.0 * 1e-6 #depth from TOI catalogue with normalised units (ppm -> normalised units)

tic_id = 276376289 
ra_dec_str = '20:53:57.48 +42:44:36.6'

# Now plot
%matplotlib inline
plt.figure()
plt.scatter(time,flux, c='k', s=1)
plt.show()
```

```python
%matplotlib inline

#Getting the aliases for the system

Pmin = 13 # This value needs to be set by you, and is the minimum period possible set by the maximum continuos data span.
          # Usually, this is 17 days.    

Nmax = int(np.floor(Pmax/Pmin))
print('Number of aliases : {:}'.format(Nmax))

# This is your alias mask
# Set indexes to False if you do not want to include the alias
# It is the P number -1
# Here, I do not like aliases 1 to 35 and 37 till the end.

alias_mask = np.ones(Nmax, dtype=bool)

alias_mask[:35] = False
alias_mask[36:] = False

# Plotting the aliases, data that folds over and overlaps are unusable aliases. 

# In this case, alias 36 is the solution as this system had 
# 3 transits and all three match with this orbital period.
# ALIAS 36 MOST PROBABLY TRUE PERIOD!!!

fig, ax = plt.subplots(Nmax, 1, figsize=(3,2*Nmax))
for i in range(1, Nmax+1):
    period = Pmax/i
    pwidth = width/period
    phase = (((time - t_zero_1)/period)+0.5)%1 - 0.5
    mask = (phase > (-pwidth)) & (phase < (pwidth))
    
    ax[i-1].errorbar(phase[mask], flux[mask], yerr=flux_err[mask], fmt='k.', markersize=1)
    ax[i-1].set(xlabel='Phase', ylabel='Flux', title='Alias P{:}\nPeriod = {:.4f}'.format(i, period), xlim=(-pwidth,pwidth))
    if not alias_mask[i-1] : ax[i-1].set_facecolor('xkcd:salmon')
plt.tight_layout()
plt.show()
```

```python
#Creating a file that can be used to create an observation schedule.

aliases = np.arange(1, Nmax+1)[alias_mask]
periods = Pmax/aliases

f = open('TIC{:}_aliases.txt'.format(tic_id), 'w+')
f.write(ra_dec_str + '\n')
f.write('PALIAS,EPOCH,EPOCH_ERR_HIGH,EPOCH_ERR_LOW,PERIOD,PERIOD_ERR_HIGH,PERIOD_ERR,DURATION')

for i in range(len(periods)):
    line = '\n{:},{:},{:},{:},{:},{:},{:},{:}'.format(aliases[i], t_zero_1 + 2457000, 0.0001,0.0001, periods[i],0.0001,0.0001, width)
    f.write(line)
f.close()
```
